```python
import pandas as pd
import numpy as np
import collections
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
import datetime as dt
from time import strftime
import pytz
import random


#show charts inline
%matplotlib inline 
```

### Set Parameters

```python
intervalLength = 15 #Aggregate readings to the nearest xxx minutes
windMin = 17
startTimeFeature = dt.datetime(2015,1,1,7,30,0) 
endTimeFeature = dt.datetime(2015,1,1,14,0,0)
startTimeTarget = dt.datetime(2015,1,1,15,0,0)
endTimeTarget = dt.datetime(2015,1,1,18,0,0)
startForecast = dt.datetime(2015,1,1,9,0,0)
endForecast = dt.datetime(2015,1,1,12,0,0)

targetType = 'regression' #binary or regression
incGustLull = True #include gust and lull data as feature
takeDiff = True #include the first difference as a feature
trainingSize = 0.7 #% of the data set allocated to training


#TO DO Add Assert statements
```

### Read Data

```python
dfWeatherWaddell = pd.read_csv('../RAW/Waddell_from_2005.csv',parse_dates=[1],infer_datetime_format=True,na_values=['FALSE'])
```

### Clean data

```python
#remove mph from column names (for convenience later)
dfWeatherWaddell['lull'] = dfWeatherWaddell['lull(mph)']
del(dfWeatherWaddell['lull(mph)'])
dfWeatherWaddell['avg'] = dfWeatherWaddell['avg(mph)']
del(dfWeatherWaddell['avg(mph)'])
dfWeatherWaddell['gust'] = dfWeatherWaddell['gust(mph)']
del(dfWeatherWaddell['gust(mph)'])

#make the timestamp the index - makes later joining easuer
dfWeatherWaddell.index = dfWeatherWaddell.timestamp
del(dfWeatherWaddell['timestamp'])
```

```python
#replace 0s with nulls
dfWeatherWaddell['lull'] = dfWeatherWaddell['lull'].replace(0, np.nan)
dfWeatherWaddell['avg'] = dfWeatherWaddell['avg'].replace(0, np.nan)
dfWeatherWaddell['gust'] = dfWeatherWaddell['gust'].replace(0, np.nan)


```

```python
#average the observations into 15 minute blocks so that we have a consistently spaced time series
nearestXXX = intervalLength*60*1000000000   # 15 minutes in nanoseconds 
dfWeatherWaddell['timestamp_nearest_xxx'] = pd.DatetimeIndex(((dfWeatherWaddell.index.astype(np.int64) // nearestXXX + 1 ) * nearestXXX))

#convert from utc to PT
pacific = pytz.timezone('US/Pacific')
dfWeatherWaddell.index = dfWeatherWaddell.index.tz_localize(pytz.utc).tz_convert(pacific)
dfWeatherWaddell['timestamp_nearest_xxx'] = pd.DatetimeIndex(dfWeatherWaddell['timestamp_nearest_xxx']).tz_localize(pytz.utc).tz_convert(pacific)
```

```python
#create a data frame using using the consistently spaced time series
def groupByEvenSpace(dfRaw,Series,intervalLength):
    dfWeatherGroupBy = dfRaw.groupby(dfRaw['timestamp_nearest_xxx'])[Series].agg(np.mean) 
    dfWeatherGroupBy = dfWeatherGroupBy.reindex(pd.date_range(min(dfWeatherGroupBy.index), max(dfWeatherGroupBy.index), freq=str(intervalLength) + 'Min'))
    dfWeatherGroupBy = dfWeatherGroupBy[(dfWeatherGroupBy.index.month >= 4) & (dfWeatherGroupBy.index.month <= 10)]
    return dfWeatherGroupBy
```

### Explore data

```python
#plot the wind by hour - sanity check
dfWeatherGroupByEvenSpace = groupByEvenSpace(dfWeatherWaddell,'avg',intervalLength)

fig, ax = plt.subplots()
fig.set_figwidth(15)
dfWeatherGroupByEvenSpace.groupby(dfWeatherGroupByEvenSpace.index.hour).agg(np.mean).plot(kind='bar')
```

```python
def turnIntoFeatureMatrix(startTime,endTime,dfWind,appendStr,takeDiff):
    timeIter = startTime
    dfFeature = pd.DataFrame(index=dfWind[(dfWind.index.hour == 17) & (dfWind.index.minute == 0)].index.date, columns=[])
    while timeIter < endTime:
        timeIter+=dt.timedelta(minutes=intervalLength)
        dfFeature[str(appendStr) + str(timeIter.hour) + "{0:0=2d}".format(timeIter.minute)] = dfWind[(dfWind.index.hour == timeIter.hour) & (dfWind.index.minute == timeIter.minute)].get_values()
        if ((takeDiff == True) and (timeIter > startTime + dt.timedelta(minutes=intervalLength))): #do take the first difference if not the first iteration
            dfFeature['diff' + str(appendStr) + str(timeIter.hour) + "{0:0=2d}".format(timeIter.minute)] = dfFeature[str(appendStr) + str(timeIter.hour) + "{0:0=2d}".format(timeIter.minute)] - dfFeature[str(appendStr) + str((timeIter-dt.timedelta(minutes=intervalLength)).hour) + "{0:0=2d}".format((timeIter-dt.timedelta(minutes=intervalLength)).minute)]
          
    return dfFeature

```

```python
dfFeatureMatrixComplete =  turnIntoFeatureMatrix(startTimeFeature,endTimeTarget,dfWeatherGroupByEvenSpace,'',False)
dfFeatureMatrixComplete['Target'] = (turnIntoFeatureMatrix(startTimeTarget,endTimeTarget,dfWeatherGroupByEvenSpace,'',False) > windMin).count(axis=1)-(turnIntoFeatureMatrix(startTimeTarget,endTimeTarget,dfWeatherGroupByEvenSpace,'',False) > windMin).sum(axis=1)<2

#plot examples of days that are kiteable
dfFeatureMatrixForPlotting = dfFeatureMatrixComplete[dfFeatureMatrixComplete['Target'] == 1]
fig, ax = plt.subplots()
fig.set_figwidth(15)
fig.suptitle('Kiteable days')
sample_of_dates = random.sample(np.unique(dfFeatureMatrixForPlotting.index), 20)
dfFeatureMatrixForPlotting.drop('Target', axis=1)[dfFeatureMatrixForPlotting.index.isin(sample_of_dates)].T.plot(ax=ax,legend=False)

dfFeatureMatrixForPlotting = dfFeatureMatrixComplete[dfFeatureMatrixComplete['Target'] == 0]
fig, ax = plt.subplots()
fig.set_figwidth(15)
fig.suptitle('Not kiteable days')
sample_of_dates = random.sample(np.unique(dfFeatureMatrixForPlotting.index), 20)
dfFeatureMatrixForPlotting.drop('Target', axis=1)[dfFeatureMatrixForPlotting.index.isin(sample_of_dates)].T.plot(ax=ax,legend=False)

```

```python
#% of kiteable days by month

(dfFeatureMatrixComplete.dropna().groupby(pd.DatetimeIndex(dfFeatureMatrixComplete.dropna().index).month).Target.agg(np.sum)/dfFeatureMatrixComplete.dropna().groupby(pd.DatetimeIndex(dfFeatureMatrixComplete.dropna().index).month).Target.count()).plot(kind='bar')
```

### Create Feature Matrix

```python
dfWeatherGroupByEvenSpace = groupByEvenSpace(dfWeatherWaddell,'avg',intervalLength)
dfFeatureMatrix = turnIntoFeatureMatrix(startTimeFeature,endTimeFeature,dfWeatherGroupByEvenSpace,'',takeDiff)

#defines a kiteable day as any day with all readings except 1 above 17 mph
dfFeatureMatrix[(dfFeatureMatrix.isnull().sum(axis=1) > 0) & (dfFeatureMatrix.isnull().sum(axis=1) < 6)] =  dfFeatureMatrix[(dfFeatureMatrix.isnull().sum(axis=1) > 0) & (dfFeatureMatrix.isnull().sum(axis=1) < 4)].astype(float).interpolate(axis=1,limit=3)
#dfFeatureMatrix.isnull().sum(axis=1).hist()

if incGustLull:
    dfWeatherGroupByEvenSpaceGust = groupByEvenSpace(dfWeatherWaddell,'gust',intervalLength)
    dfFeatureMatrixGust = turnIntoFeatureMatrix(startTimeFeature,endTimeFeature,dfWeatherGroupByEvenSpaceGust,'gust',False)
    dfFeatureMatrixGust[(dfFeatureMatrixGust.isnull().sum(axis=1) > 0) & (dfFeatureMatrixGust.isnull().sum(axis=1) < 5)] =  dfFeatureMatrixGust[(dfFeatureMatrixGust.isnull().sum(axis=1) > 0) & (dfFeatureMatrixGust.isnull().sum(axis=1) < 5)].astype(float).interpolate(axis=1,limit=2)
    dfFeatureMatrixGust.isnull().sum(axis=1).hist()
    dfFeatureMatrix = dfFeatureMatrix.join(dfFeatureMatrixGust)

    dfWeatherGroupByEvenSpaceLull = groupByEvenSpace(dfWeatherWaddell,'lull',intervalLength)
    dfFeatureMatrixLull = turnIntoFeatureMatrix(startTimeFeature,endTimeFeature,dfWeatherGroupByEvenSpaceLull,'lull',False)
    dfFeatureMatrixLull[(dfFeatureMatrixLull.isnull().sum(axis=1) > 0) & (dfFeatureMatrixLull.isnull().sum(axis=1) < 4)] =  dfFeatureMatrixLull[(dfFeatureMatrixLull.isnull().sum(axis=1) > 0) & (dfFeatureMatrixLull.isnull().sum(axis=1) < 4)].astype(float).interpolate(axis=1,limit=2)
    dfFeatureMatrix = dfFeatureMatrix.join(dfFeatureMatrixLull)


dfFeatureMatrix = dfFeatureMatrix.dropna()
print 'This feature matrix has ' + str(len(dfFeatureMatrix)) + ' rows'
```

### Create Target

```python
def createTarget(startTimeTarget,endTimeTarget,dfWeatherGroupByEvenSpace):
    if targetType == 'binary':
        #true if every reading except 1 is greater than the wind min
        target = (turnIntoFeatureMatrix(startTimeTarget,endTimeTarget,dfWeatherGroupByEvenSpace,'') > windMin).count(axis=1)-(turnIntoFeatureMatrix(startTimeTarget,endTimeTarget,dfWeatherGroupByEvenSpace,'') > windMin).sum(axis=1)<2
    else:
        #regression target
        target = (turnIntoFeatureMatrix(startTimeTarget,endTimeTarget,dfWeatherGroupByEvenSpace,'',False)).mean(axis=1)
    return target

```

```python
dfFeatureMatrix['Target'] = createTarget(startTimeTarget,endTimeTarget,dfWeatherGroupByEvenSpace)
```

### Train Random Forest

```python
#Split between test and training
msk = np.random.rand(len(dfFeatureMatrix)) < trainingSize
train = dfFeatureMatrix[msk]
test = dfFeatureMatrix[~msk]
```

```python
def columnNames(startTime,endTime,incGustLull=False):
    timeIter = startTime
    columnFeatures = ['Target']
    while timeIter < endTime:
        timeIter+=dt.timedelta(minutes=intervalLength)
        columnFeatures.append(str(timeIter.hour) + "{0:0=2d}".format(timeIter.minute)) 
        if incGustLull:
            columnFeatures.append('gust' + str(timeIter.hour) + "{0:0=2d}".format(timeIter.minute)) 
            columnFeatures.append('lull' + str(timeIter.hour) + "{0:0=2d}".format(timeIter.minute)) 
    return columnFeatures
```

```python
def evaluateModel(output,test):
    if targetType == 'binary':
        score = np.float(sum(output == test))/np.float(len(output))
    else:
        #plt.scatter(output, test)
        score = np.corrcoef(output,test)[0][1]
        #print np.corrcoef(output,test)
    return score
```

```python
#run the models looping through different cutoffs
forest = RandomForestClassifier(n_estimators = 1000)

#first iteration is has four features 
timeIter = startForecast

#a data frame for storing the reslts
dfForecastEval = pd.DataFrame(data=None, index=columnNames(startForecast,endForecast,False)[1:], columns=['errorRate'], dtype=None, copy=False)

i = 0
while timeIter < endForecast:
    

    timeIter+=dt.timedelta(minutes=intervalLength)
    
    trainTmp = train[columnNames(startTimeFeature, timeIter,False)]
    #trainTmp['Vector'] = trainTmp[str(timeIter.hour) + "{0:0=2d}".format(timeIter.minute)] - trainTmp[str((startTimeFeature+dt.timedelta(minutes=(intervalLength*2))).hour) + "{0:0=2d}".format((startTimeFeature+dt.timedelta(minutes=(intervalLength*2))).minute) ]
    #trainTmp['Month'] = pd.DatetimeIndex(trainTmp.index).month
    
    testTmp = test[columnNames(startTimeFeature, timeIter,False)]
    #testTmp['Vector'] = testTmp[str(timeIter.hour) + "{0:0=2d}".format(timeIter.minute)] - testTmp[str((startTimeFeature+dt.timedelta(minutes=(intervalLength*2))).hour) + "{0:0=2d}".format((startTimeFeature+dt.timedelta(minutes=(intervalLength*2))).minute) ]
    #testTmp['Month'] = pd.DatetimeIndex(testTmp.index).month
    
    forest = forest.fit(trainTmp.drop('Target', axis=1),train['Target'])
    output = forest.predict(testTmp.drop('Target', axis=1))
    dfForecastEval['errorRate'][i] = evaluateModel(output,testTmp['Target'].values)
    print str(dfForecastEval.index[i]) + ' ' + str(dfForecastEval['errorRate'][i])

    
    forest.feature_importances_
    dfImportances = pd.DataFrame(index=trainTmp.drop('Target', axis=1).columns)
    dfImportances['importances'] = forest.feature_importances_
    #print dfImportances
    
    
    i+=1
    
```

```python
#plot the correlation between predicted and actual
fig, ax = plt.subplots()
fig.set_figwidth(15)

print dfForecastEval['1100':'1200'].mean()
dfForecastEval.plot(kind='bar',ax=ax)
```
